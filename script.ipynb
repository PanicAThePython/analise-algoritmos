```python

recursive_method_used_items = []
class Item:
    def __init__(self, value, weight):
        self.value = value
        self.weight = weight

#items = Item[]

def mochila_inteira_bottom_up(n, items, W):
    m = [[0]*(W+1) for i in range(n+1)]
    
    for x in range(0, W):
        m[0][x] = 0
    
    for j in range(0, n):
        m[j][0] = 0
        
    for j in range(1, n+1):
        wj = items[j-1].weight
        vj = items[j-1].value
        
        for x in range(0, W+1):
            if wj > x:
                m[j][x] = m[j-1][x]
            else:
                use = vj + m[j-1][x-wj]
                dont_use = m[j-1][x]
                m[j][x] = max(use, dont_use)
            
    #return m[n][W]
    return get_ideal_arrangement(n, m, items, W)

def mochila_inteira_recursive_2(n, items, W):
    global recursive_method_used_items
    if n == 0 or W == 0:
        return 0
    
    wn = items[n-1].weight
    
    if wn > W:
        return mochila_inteira_recursive_2(n-1, items, W)
    else:
        vn = items[n-1].value
        use = vn + mochila_inteira_recursive_2(n-1, items, W - wn)
        dont_use = mochila_inteira_recursive_2(n-1, items, W)
        
        result = max(use, dont_use)
        
        if use > dont_use:
            can_add = True
            for item in recursive_method_used_items:
                if item == n:
                    can_add = False
                    break
            if can_add:
                recursive_method_used_items.append(n)
            
        return result

def mochila_inteira_recursive(n, items, W):
    global recursive_method_used_items
    recursive_method_used_items = []
    formatted_recursive_method_used_items = []
    mochila_inteira_recursive_2(n, items, W)
    recursive_method_used_items = sorted(recursive_method_used_items, key=lambda x: x, reverse=True)
    sum = 0
    for i in recursive_method_used_items:
        if sum >= W or sum + items[i-1].weight > W:
            return formatted_recursive_method_used_items
        formatted_recursive_method_used_items.append(i)
        sum += items[i-1].weight
        
        
def get_ideal_arrangement(n, m, items, W):
    used_items = []
    
    while n >= 1:
        if m[n][W] == m[n-1][W-items[n-1].weight] + items[n-1].value:
            used_items.append(n)
            W -= items[n-1].weight
        n -= 1
    return used_items

import os
from time import time


# items = []
    
# items.append(Item(1, 1))
# items.append(Item(5, 2))
# items.append(Item(8, 4))
# items.append(Item(11, 5))
# items.append(Item(23, 6))
    
def read_backpack_from_file(file_name: str):
    file_handle = open(file_name, 'r')
    lines = file_handle.read().split('\n')

    items = []
    backpack_total_size = 0
    for line in lines:
        if " " in line:
            item_value = int(line.split(" ")[0])
            item_weight = int(line.split(" ")[1])
            items.append(Item(item_value, item_weight))
        else:
            backpack_total_size = int(line)
        
    
    file_handle.close()
    return items, backpack_total_size
        
def print_used_items(items, used_items_indexes):
    for index in used_items_indexes:
        print('Foi usado o item de valor {0} de peso {1}'.format(items[index-1].value, items[index-1].weight))

def main():
    for file in os.listdir('files'):
        if not file.endswith('.txt'):
            continue
        
        
        items, backpack_total_size = read_backpack_from_file('files/{0}'.format(file))
        print('Resultado do arquivo: {0}'.format(file))
    
    #result = mochila_inteira_bottom_up(3, items, 6)
        start_time = time()
        used_items = mochila_inteira_bottom_up(len(items), items, backpack_total_size)
        print_used_items(items, used_items)
        end_time = time()
        print('The operation using bottomUp method took {0} ms to finish'.format(end_time - start_time))
    
        start_time = time()
        used_items = mochila_inteira_recursive(len(items), items, backpack_total_size)
        end_time = time()
        print('The operation using recursive method took {0} ms to finish\n'.format(end_time - start_time))
    
    
    
    
if __name__ == "__main__":
    main()
            
```
